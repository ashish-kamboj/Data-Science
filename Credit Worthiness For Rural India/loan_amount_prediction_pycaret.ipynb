## Data Pre-processing

```python
## 'social_class' column cleaning and standardization

sc_list = ['SC','S.C','S.C.','Sc','Sechudle Caste','S.c','Schudle Cast','sc','s.c','Scheduled Caste','s.c.','S C','Schduled Cast','St','Sechudle Cast',
           'Sechudle caste','Harijan_SC','Scedelt Cast','Mahar(SC)','Schedule Caste','Chambhar-SC','chamar','st','Hindu_SC','Chambar','Harijan','Sechudled Caste',
           'Chamar','Chambhar','Chabhar','Valmiki','Chambahr','Prajapati']

st_list = ['ST','S.T.','S.T','Sechudle Tripe','Scheduled tribe','Scheduled Tribe','S. t.','ST Gond']
gen_list = ['General','G.C','GEN','General Caste','Gen','GENERAL','GENL','Genarel','Genaral','gen','Genl','Genral','general','Gen.','General caste',
            'Genarel Cast','GENERALH','GEN.','Genetal','Gen- BPL','Genrel','genral','Geberal','GC']
obc_list = ['OBC','Obc','O.B.C','O.B.C.','obc','O B C','Other Backward caste','Hindu_OBC','0bc','Sonar (OBC)','Kasar- OBC','OBC`','OC']
hindu_list = ['Hindu','Hindhu','HINDU','Hindu, Kuruba','hindu','Hindu_Ezhava','Hindu, Waddar']
muslim_list = ['Muslim','MUSLIM','Musilim','Mushlim','Islam','muslim','Mulim','Muslin','Musilm','Muslimm']
minority_list = ['MINORITY','Minority Comunity','Mynority','Mynirity','Minerity','mionrity','Minerty','Minorty','minority','Minority comunity','Mynorety',
                'Maniroty Comunity']
nomadic_list = ['NT','N.T.','Dhangar(Nomadic Tribes)','Nt','TRIBEL','Dhangar(Nomadic Tribe)','Dhangar','Dhangar(Nomadic tribes)','banjaran','Tribal']
open_list = ['Open','OPEN','open','Opan']
lingayat_list = ['Lingahitha','Lingayath','lingayat','Linghaitha','Lingyatha','Lingayat  Vani']
christian_list = ['Christian','christian','Chricshan','Christain','Crichtion','Cristan']
buddha_list = ['Bauddha','Nava-Baudha','Bauddh','Buddha','Baudhha','Baudh','Bauoddh']
backward_list = ['BC','MBC','BCE','BC-E','SEBC','BC-B']


## Function for standardizing the 'social_class' column
def social_class_standardization(x):
    
    if(x in sc_list):
        return 'SC'
    elif(x in st_list):
        return 'ST'
    elif(x in gen_list):
        return 'General'
    elif(x in obc_list):
        return 'OBC'
    elif(x in hindu_list):
        return 'Hindu'
    elif(x in muslim_list):
        return 'Muslim'
    elif(x in minority_list):
        return 'Minority'
    elif(x in nomadic_list):
        return 'Nomadic Tribe'
    elif(x in open_list):
        return 'Open'
    elif(x in lingayat_list):
        return 'Lingayat'
    elif(x in christian_list):
        return 'Christian'
    elif(x in buddha_list):
        return 'Buddha'
    elif(x in backward_list):
        return 'Backward Class'
    else:
        return x
        
df_cleaned = df.copy()
df_cleaned['social_class'] = df_cleaned['social_class'].apply(lambda x: social_class_standardization(x))

## Converting the 'city' names to titlecase
df_cleaned['city'] = df_cleaned['city'].str.title()

#df_subset['occupants_count'] = df_subset['occupants_count'].apply(lambda x: x if(x<11) else 10)
```

```python
## Filtering Data

df_cleaned = df_cleaned[df_cleaned['age'] <= 140]
df_cleaned = df_cleaned[df_cleaned['annual_income'] > 0]
df_cleaned = df_cleaned[df_cleaned['monthly_expenses'] < df_cleaned['annual_income']]
df_cleaned = df_cleaned[df_cleaned['occupants_count'] <= 10]
df_cleaned = df_cleaned[df_cleaned['house_area'] > 0]
```

```python
## Function for checking percentage of null values in dataframe columns

def null_value_check_in_dataframe(df):
    percent_missing = df.isnull().sum() * 100 / len(df)
    missing_value_df = pd.DataFrame({'column_name': df.columns,
                                     'percent_missing': percent_missing})
    return missing_value_df

df_null = null_value_check_in_dataframe(df_cleaned)
df_null
```
